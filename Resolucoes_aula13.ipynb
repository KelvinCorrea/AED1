```python
# Entrada do usuário
frase = input("Digite uma frase: ")

# Transforma a frase em minúsculas
frase = frase.lower()

# Remove os espaços em excesso
frase = ' '.join(frase.split())

# Substitui os espaços restantes por underscores
frase = frase.replace(' ', '_')

# Exibe o resultado
print(F"Frase normalizada: {frase}")
```

```python
# Exercício 2

# Lista pré-definida de vendas diárias
vendas = [50, 150, 200, 80, 300, 90, 120]

# Remove as vendas menores que R$ 100,00
vendas_filtradas = [venda for venda in vendas if venda >= 100]

# Calcula a média das vendas restantes, se houver
if len(vendas_filtradas) > 0:
    media_vendas = sum(vendas_filtradas) / len(vendas_filtradas)
    print("A média das vendas restantes é: R$", round(media_vendas, 2))
else:
    print("Nenhuma venda acima de R$ 100,00.")
```

```python
# Exercício 3

import math

# Recebe as três coordenadas (x, y) como tuplas
ponto1 = tuple(map(float, input("Digite as coordenadas do ponto 1 (x, y) separadas por espaço: ").split()))
ponto2 = tuple(map(float, input("Digite as coordenadas do ponto 2 (x, y) separadas por espaço: ").split()))
ponto3 = tuple(map(float, input("Digite as coordenadas do ponto 3 (x, y) separadas por espaço: ").split()))

# Calcula a distância Euclidiana de cada ponto até a origem (0, 0)
distancia1 = math.sqrt(ponto1[0]**2 + ponto1[1]**2)
distancia2 = math.sqrt(ponto2[0]**2 + ponto2[1]**2)
distancia3 = math.sqrt(ponto3[0]**2 + ponto3[1]**2)

# Calcula a distância média
media_distancia = (distancia1 + distancia2 + distancia3) / 3

# Exibe o resultado
print("A distância média dos pontos a partir da origem é:", round(media_distancia, 2))
```

```python
# Exercício 4

# Recebe a frase de entrada
frase = input("Digite uma frase: ")

# Transforma a frase em minúsculas e divide em palavras
palavras = frase.lower().split()

# Cria um dicionário para contar a frequência de cada palavra
frequencia = {}

# Conta a frequência de cada palavra
for palavra in palavras:
    if palavra in frequencia:
        frequencia[palavra] += 1
    else:
        frequencia[palavra] = 1

# Exibe a contagem de cada palavra
print("Frequência de palavras:")
for palavra, contagem in frequencia.items():
    print(f"{palavra}: {contagem}")
```
